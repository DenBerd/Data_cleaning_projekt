3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Функция извлечения опыта работы в месяцах
def get_job_xp(experience):
    
    if experience == "Не указано" or experience is np.nan:
        return (np.nan)
    else:
        xp_years = 0
        xp_months = 0
        xp_list = str(experience).split()[:9]
        for i in range((len(xp_list))-1):
            if xp_list[i].isdigit() and (xp_list[i+1].startswith("лет") or xp_list[i+1].startswith("год")):
                xp_years += int(xp_list[i])
                break
        for i in range((len(xp_list))-1):
            if xp_list[i].isdigit() and xp_list[i+1].startswith("месяц"):
                xp_months += int(xp_list[i])
                break
        if xp_years > 100:
            xp_years = 0        
        return xp_years*12 + xp_months


# Применяем функцию    
hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(get_job_xp)
# Контрольный вопрос
print(hh_data['Опыт работы (месяц)'].max())
```

```python
hh_data.drop('Опыт работы', axis = 1, inplace=True)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#Функция извлечения общей информации о городе
def general_city(city_cell):
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск',
                      'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь',
                      'Воронеж', 'Волгоград' ]
    city = city_cell.split(',')[0].strip()
    if city in million_cities:
        return 'город-миллионник'
    elif city == 'Москва' or city == 'Санкт-Петербург':
        return city
    else:
        return 'другие'
# Применяем функцию   
hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(general_city)

#Функция получения информации о готовности к переезду
def ready_to_move(city_cell):
    info_list = city_cell.split(',')[1:]
    for i in info_list:
        if i.strip().startswith('м.'):
            continue
        elif i.strip().startswith('не гот'):
            return False
        else:
            return True
# Применяем функцию
hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(ready_to_move) 

#Функция получения информации по командировкам
 
def ready_for_trips(city_cell):
    info_list = city_cell.split(',')
    last_element = info_list[-1].strip()
    ready = last_element.startswith('готов') and 'командир' in last_element
    return ready

hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(ready_for_trips)

#контрольный вопрос
#two_for_one = hh_data['Готовность к переезду']&hh_data['Готовность к командировкам']
#two_for_one.value_counts(normalize = True)
hh_data['Город'].value_counts(normalize = True)
```

```python
#ячейка удаления колонки
hh_data.drop(['Город, переезд, командировки'], axis = 1, inplace=True)
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
import re
#загружаем датафрейм с данными валют
exchange_df = pd.read_csv('data/ExchangeRates.csv')

#выполняем подготовку таблиц к слиянию
exchange_df['date'] = pd.to_datetime(exchange_df['date'])
exchange_df['date']=exchange_df['date'].dt.date
hh_data['Обновление резюме'] = pd.to_datetime(hh_data['Обновление резюме'])
hh_data['Обновление резюме'] = hh_data['Обновление резюме'].dt.date
hh_data['salary'] = hh_data['ЗП'].apply(lambda x: int(re.search('\d+', x).group()))

#унифицируем наименования валют
def salary_naming(cell):
    cell = re.search('\D+', cell).group().strip()
    if cell == 'руб.':
        return 'RUR'
    elif cell == 'грн.':
        return 'UAH'
    elif cell == 'сум':
        return 'UZS'
    elif cell == 'бел.руб.':
        return 'BYN'
    else:
        return cell
    
hh_data['currency'] = hh_data['ЗП'].apply(salary_naming)

#выполняем слияние
hh_data = hh_data.merge(
         exchange_df,
         left_on=['Обновление резюме', 'currency'],
         right_on=['date', 'currency'],
         how='left'
)

#заполняем пропуски
hh_data = hh_data.fillna({'close': 1.0, 'proportion': 1.0})
#добавляем итоговую колонку
hh_data['ЗП (руб.)'] = hh_data['salary'] * hh_data['close'] / hh_data['proportion']


#контрольный вопрос
hh_data['ЗП (руб.)'].median()
```

```python
#ячейка удаления колонок
hh_data = hh_data.drop(['ЗП', 'salary', 'currency', 'per',
                           'date', 'time', 'close', 'vol', 'proportion'], axis = 1)
```

По графику видно, что возраст смещен вправо, то есть людей старшего возраста больше, чем аномально молодых.

```python
#вспомним полезную функцию
def outliers_z_score(data, column, log_scale=False):
    if log_scale:
        sequence = np.log(data[column]+1)
    else:
        sequence = data[column]
    mu = sequence.mean()
    sigma = sequence.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers = data[(sequence < lower_bound) | (sequence > upper_bound)]
    normal_data = data[(sequence > lower_bound) & (sequence < upper_bound)]
    return outliers, normal_data

outliers, normal_age = outliers_z_score(hh_data, 'Возраст', log_scale=True)
print(outliers)
```
